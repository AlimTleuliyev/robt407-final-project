```python
import os
from glob import glob
import pandas as pd
```

```python
training_csvs = glob('experiments/**/*metrics.csv', recursive=True)
training_csvs
```

```python
temp_df = pd.read_csv(training_csvs[0])
temp_df
```

```python

# we need to get best row for every model and create new dataframe with best rows for every model
best_rows = []
for csv in training_csvs:
    model_name = os.path.basename(csv).split('_training_metrics')[0]
    df = pd.read_csv(csv)
    df['Model'] = model_name
    best_row = df.loc[df['Val F1'].idxmax()]
    # add model name to best row which is a series
    best_rows.append(best_row)

best_df = pd.DataFrame(best_rows)
best_df = best_df.sort_values(by=['Val F1'], ascending=False)
best_df.to_csv('best_models_results.csv', index=False)
best_df
```
